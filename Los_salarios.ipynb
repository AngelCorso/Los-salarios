## Transformación de datos

Con el propósito de preparar los datos para el entrenamiento de un modelo de inteligencia artificial, se decidió utilizar las técnicas de Ordinal Encoding y One Hot Encoding en las variables categóricas.

### Ordinal Encoding

```python

from sklearn.preprocessing import OrdinalEncoder

# ordinales: remote_ratio, experience_level, company_size

enc = OrdinalEncoder()
enc.fit(df[["experience_level","company_size"]])
df_transformed = pd.DataFrame()
# df_transformed[['salary_in_usd']] = df[['salary_in_usd']]
df_transformed[["experience_level","company_size"]] = enc.transform(df[["experience_level","company_size"]])

df_transformed['remote_ratio'] = df['remote_ratio']
# df[["experience_level","company_size"]] = enc.transform(df[["experience_level","company_size"]])
df_transformed.count()
```

Se decidió eliminar lo registros duplicados.

```python
df_transformed = df_transformed.loc[df_transformed.duplicated() == False]
```

Se decidieron eliminar los valores atípicos tras identificar valores extraños con base en el contexto del problema.

```python
std_value= df_transformed["salary_in_usd"].std()
df_transformed = df_transformed.loc[np.abs((df_transformed["salary_in_usd"] - df_transformed["salary_in_usd"].mean()) / std_value)  < 3]
```
